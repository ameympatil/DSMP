### Colored Scatterplots

```python
iris = pd.read_csv('/content/iris.csv')
iris.head()
```

```python
iris['Species']=iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],alpha=0.5)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### Annotations

```python
batters = pd.read_csv('batter.csv')
batters.head()
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
```

```python
sample_df
```

```python
plt.figure(figsize=(18,12))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
x=[1,2,3,4]
y=[5,6,7,8]

plt.scatter(x,y)
plt.text(1,5,'Point 1')
plt.text(2,6,'Point 2')
plt.text(3,7,'Point 3')
plt.text(4,8,'Point 4')
```

### Horizontal and Vertical lines

```python
plt.figure(figsize=(18,12))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red')
plt.axhline(140,color='green')
plt.axvline(30,color='red')

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
figure,ax = plt.subplots(nrows=2,ncols=1,figsize=(8,8),sharex=True)


ax[0].scatter(batters['avg'],batters['strike_rate'])
ax[0].set_title('Avg vs Strike Rate')
ax[0].set_ylabel('Strike Rate')
ax[0].set_xlabel('Avg')

ax[1].scatter(batters['avg'],batters['runs'])
ax[1].set_ylabel('Runs')
ax[1].set_title('Avg vs Runs')
ax[1].set_xlabel('Avg')

figure.show()
```

```python
figure,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))


ax[0,0].scatter(batters['avg'],batters['strike_rate'])
ax[0,0].set_title('Avg vs Strike Rate')
ax[0,0].set_ylabel('Strike Rate')
ax[0,0].set_xlabel('Avg')

ax[0,1].scatter(batters['avg'],batters['runs'])
ax[0,1].set_ylabel('Runs')
ax[0,1].set_title('Avg vs Runs')
ax[0,1].set_xlabel('Avg')

ax[1,0].hist(batters['avg'])
ax[1,0].set_title('Avg')

ax[1,1].hist(batters['runs'])
ax[1,1].set_title('Runs')

figure.show()
```

### 3D Line Plot

```python
x = [0,1,5,25]
y = [0,10,13,0]
z = [0,13,20,9]


fig = plt.figure()

ax = plt.subplot(projection='3d')

ax.scatter3D(x,y,z,s=[100,100,100,100])
ax.plot3D(x,y,z,color='red')
```

### 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx,yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
fig.show()
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
fig.show()
```

```python
z = np.sin(xx) + np.log(xx)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
fig.show()
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
fig.show()
```

### Contour Plots

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
fig.show()
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
fig.show()
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
fig.show()
```
