# Scatter Plots

image.png

- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)
x

y = 10*x + 3+ np.random.randint(0,300,50)
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data
df = pd.read_csv('/content/batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'])
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.title('AR & SR Comparison of top 50 batsmen')
```

```python
# marker
plt.scatter(df['avg'],df['strike_rate'],marker='+',color='red')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.title('AR & SR Comparison of top 50 batsmen')
```

```python
# size
tips = sns.load_dataset('tips')
tips

plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*10)
```

```python
# scatterplot using plt.plot
plt.plot(tips['total_bill'],tips['tip'],'o')
```

```python
# plt.plot vs plt.scatter
# plt.plot is faster than plt.scatter
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='green')
```

```python
# bar chart using data
```

```python
# horizontal bar chart
plt.barh(colors,children,color='green')
```

```python
# color and label
df = pd.read_csv('/content/batsman_season_record.csv')
df

plt.bar(df['batsman'],df['2015'])
```

```python
# Multiple Bar charts
plt.bar(np.arange(df.shape[0])-0.25,df['2015'],width=.15,label='2015')
plt.bar(np.arange(df.shape[0]),df['2016'],width=.15,label='2016')
plt.bar(np.arange(df.shape[0])+0.25,df['2017'],width=.15,label='2017')

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.legend()
plt.show()
```

```python
# xticks
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red red','blue blue blue blue blue','green green green green green','yellow yellow yellow yellow yellow','pink pink pink pink pink']

plt.bar(colors,children)
plt.xticks(rotation=70)
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'])
plt.bar(df['batsman'],df['2016'],bottom=df['2017'])
plt.bar(df['batsman'],df['2015'],bottom=df['2016'] + df['2017'])
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data

data = [23,45,100,20,49]
subjects = ['eng','sci','maths','sst','hindi']

plt.pie(data,labels = subjects)
plt.show()
```

```python
# dataset
df = pd.read_csv('/content/gayle-175.csv')
df.head()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.5,0,0,0,0,0],shadow=True)
plt.show()
```
